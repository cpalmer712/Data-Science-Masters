```python
import numpy as np
import matplotlib.pyplot as plt

# Wiggly Function Definition
def f_2(x):
    return -np.exp(-x**2 / 100) * np.sin(13*x - x**4 / 5) * np.sin(1 - 3*x**2) / 2

# Simulated Annealing Function
def sim_anneal(f, x, x_bounds, temp, rate):
    def random_neighbour(x, bounds):
        new_x = x + np.random.uniform(-0.1, 0.1) * (bounds[1] - bounds[0])
        return np.clip(new_x, *bounds)
    
    def acceptance_probability(old_cost, new_cost, temp):
        if new_cost < old_cost:
            return 1
        else:
            return np.exp((old_cost - new_cost) / temp)
    
    current_x = x
    best_x = current_x
    path = [current_x]
    while temp > 0.01:
        new_x = random_neighbour(current_x, x_bounds)
        current_cost = f(current_x)
        new_cost = f(new_x)
        if acceptance_probability(current_cost, new_cost, temp) > np.random.random():
            current_x = new_x
            if new_cost < f(best_x):
                best_x = new_x
                path.append(best_x)
        temp *= rate
    return np.array(path)

# Finding Minimum of the Wiggly Function
initial_guess = 0
x_bounds = (-10, 10)
initial_temp = 1000
temp_rate = 0.99
solution_path = sim_anneal(f_2, initial_guess, x_bounds, initial_temp, temp_rate)
best_solution = solution_path[-1]

# Printing the Best Solution Found
print("Best solution found:", best_solution)

# Plotting the Solution Path
x_values = np.linspace(*x_bounds, 1000)
plt.plot(x_values, f_2(x_values), label="Wiggly Function")
plt.scatter(solution_path, f_2(solution_path), color='red', label="Solution Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Minimization of Wiggly Function using Simulated Annealing")
plt.legend()
plt.show()
```
